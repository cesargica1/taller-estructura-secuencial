```python
# Escriba una funci贸n que reciba el nombre de un departamento y que retorne cu谩ntos municipios hay
# en el departamento con ese nombre.
def municipios_departamento(muns: pd.DataFrame, nombre: str) -> int:
    n = muns[muns["departamento"]== nombre].shape[0]
    return n

# Pruebas de la funci贸n anterior
assert municipios_departamento(muns, "Cundinamarca") == 117
assert municipios_departamento(muns, "Boyac谩") == 123
assert municipios_departamento(muns, "Vichada") == 4
print("Prueba superada ")
```

```python
# Escriba una expresi贸n que permita encontrar los nombres de los municipios que tienen
# menos de 500 habitantes en la poblaci贸n rural
m = muns[muns["poblaci贸nRural"] < 500]["nombre"]

# Prueba
assert m.sort_values().to_list() == ['La Guadalupe', 'Santa Luc铆a', 'Soplaviento', 'Suan']
print("Prueba superada ")
```

```python
# Cual es la suma total de la poblaci贸n urbana de todos los municipios
pob_urb = muns["poblaci贸nUrbana"].sum()

# Prueba
assert pob_urb == 37_740_751
print("Prueba superada ")
```

```python
# Hallar la suma de la poblaci贸n urbana de los municipios de los departamentos
# del eje cafetero: Caldas, Quindio o Risaralda
pob_eje_cafetero = muns[muns["departamento"].isin(["Quind铆o", "Risaralda","Caldas"])]["poblaci贸nUrbana"].sum()

# Prueba
assert pob_eje_cafetero == 1_471_055
print("Prueba superada ")
```

```python
# Encontrar el nombre del municipio que es la capital (esCapital == 1) del departamento de Casanare
cap_casanare = muns[(muns["departamento"]=="Casanare")&(muns["esCapital"]==1)]["nombre"].iloc[0]
print(cap_casanare)
```

```python
# Encontrar cuantos municipios tienen una poblaci贸n urbana inferior al promedio
# de la poblaci贸n urbama de todo el pais
prom_poblacion_urbana = muns["poblaci贸nUrbana"].mean()
cant_muns_pobl_inf_prom = (muns["poblaci贸nUrbana"] < prom_poblacion_urbana).sum()


# Prueba
assert cant_muns_pobl_inf_prom == 998
print("Prueba superada ")
```

```python
# Presentar el nombre y problaci贸n urbana de los municipios cuya poblaci贸n urbana est茅 entre el mill贸n y los tres millones
# de habitantes.
pob_urb_alta = muns[(muns["poblaci贸nUrbana"] >= 1000000)&(muns["poblaci贸nUrbana"] <= 3000000)][["nombre", "poblaci贸nUrbana"]]
print(pob_urb_alta)
```
